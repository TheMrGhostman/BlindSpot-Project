```python
%pylab --no-import-all notebook
import pandas as pd 
from sklearn import model_selection, svm
import os 
import warnings
warnings.filterwarnings('ignore')
way = os.getcwd()+"/sgd/data/"
os.listdir()
```

# Linearly separable data 

## Inicialization

```python
os.listdir(way)
```

```python
data = pd.read_csv(way + "linsep-traindata.csv", header=None)
```

```python
labels = pd.read_csv(way + "linsep-trainclass.csv", header=None)
```

```python
plt.figure("First view on data")
plt.scatter(data[0], data[1], c = labels[0], cmap = plt.cm.plasma)
plt.show()
```

## Training and testing

```python
[data_train, data_test, labels_train, labels_test] = model_selection.train_test_split(data, labels, test_size=0.3)
```

```python
clf = svm.SVC(kernel = 'linear')
```

```python
clf.fit(data_train, labels_train)
```

```python
clf.score(data_test, labels_test)
```

```python
plt.figure("Training data and unpredicted testing data")
plt.scatter(data_train[0], data_train[1], c = labels_train[0], cmap = plt.cm.plasma, s = 10)
plt.scatter(data_test[0], data_test[1], color = "green", marker = "x", s = 100)
plt.show()
```

```python
plt.figure("Data after prediction")
plt.scatter(data_train[0], data_train[1], c = labels_train[0], cmap =plt.cm.plasma, s = 10)
plt.scatter(data_test[0], data_test[1], marker = "x", c = clf.predict(data_test), cmap = plt.cm.plasma, s = 100)
plt.show()
```

## Cross validation score

```python
scores = model_selection.cross_val_score(clf,data, labels, cv = 10)
```

```python
scores
```

# Linearly non-separable data

## Initialization

```python
data_ns = pd.read_csv(way + "nonlinsep-traindata.csv", header=None)
```

```python
labels_ns = pd.read_csv(way + "nonlinsep-trainclass.csv", header=None)
```

```python
plt.figure("First view on non-separable data")
plt.scatter(data_ns[0], data_ns[1], c = labels_ns[0], cmap = plt.cm.plasma)
plt.show()
```

```python
[data_train_ns, data_test_ns, labels_train_ns, labels_test_ns] = model_selection.train_test_split(data_ns, labels_ns, test_size=0.3)
```

```python
clf_ns = svm.SVC(kernel = 'poly')
```

```python
clf_ns.fit(data_train_ns, labels_train_ns)
```

```python
clf_ns.score(data_test_ns, labels_test_ns)
```

```python
plt.figure("Training data and unpredicted testing (non-sep) data")
plt.scatter(data_train_ns[0], data_train_ns[1], c = labels_train_ns[0], cmap = plt.cm.plasma, s = 10)
plt.scatter(data_test_ns[0], data_test_ns[1], color = "green", marker = "x", s = 100)
plt.show()
```

```python
plt.figure("(non-sepp) Data after prediction")
plt.scatter(data_train_ns[0], data_train_ns[1], c = labels_train_ns[0], cmap =plt.cm.plasma, s = 10)
plt.scatter(data_test_ns[0], data_test_ns[1], marker = "x", c = clf_ns.predict(data_test_ns), cmap = plt.cm.plasma, s = 100)
plt.show()
```

## Cross validation score

```python
scores_ns = model_selection.cross_val_score(clf_ns, data_ns, labels_ns, cv = 10)
```
