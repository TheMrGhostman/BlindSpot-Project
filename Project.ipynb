```python
%pylab --no-import-all notebook
import pandas as pd 
from sklearn import model_selection, svm
import os 
import warnings
warnings.filterwarnings('ignore')
way = os.getcwd()+"/sgd/data/"
os.listdir()
```

```python
os.listdir(way)
```

```python
data = pd.read_csv(way+"linsep-traindata.csv", header=None)
```

```python
labels = pd.read_csv(way+"linsep-trainclass.csv", header=None)
```

```python
plt.figure("NÃ¡hled")
plt.scatter(data[0],data[1], c = labels[0],cmap = plt.cm.plasma)
plt.show()
```

```python
[data_train, data_test, labels_train, labels_test] = model_selection.train_test_split(data, labels, test_size=0.3)
```

```python
clf = svm.LinearSVC()
```

```python
clf.fit(data_train, labels_train)
```

```python
clf.score(data_test,labels_test)
```

```python
plt.figure("Training data and unpredicted testing data")
plt.scatter(data_train[0], data_train[1], c = labels_train[0] , cmap = plt.cm.plasma, s = 10)
plt.scatter(data_test[0], data_test[1], color = "green" , marker = "x", s = 100)
plt.show()
```

```python
plt.figure("Data after prediction")
plt.scatter(data_train[0], data_train[1], c = labels_train[0], cmap =plt.cm.plasma, s = 10)
plt.scatter(data_test[0], data_test[1], marker = "x", c = clf.predict(data_test), cmap = plt.cm.plasma, s = 100)
plt.show()
```
